## Lambda Function

A lambda function is a small anonymous function(means it dont have any name)
A lambda function can take any number of argument, but can have one expression.

  ## Syntax:   lambda a,b : a+b

```python
# x--> x^2

lambda x: x**2
```

```python
a=lambda x: x**2
a(2)
```

```python
z=lambda x,y : x+y
z(2,3)
```

## Difference between Lambda and Normal Function

1. No name
2. Lambda function has no return value(infact it return a function)
3. lambda is written in 1 line
4. not reusable

 ## we use Lambda function? 
They are use with HOF(Higher Order Function)

```python
# check if a string has 'a'
a=lambda s : 'o' in s
a('hello')
```

```python
# odd or even
a= lambda x : x % 2==0
a(7)
```

```python
# odd or even
a= lambda x :'even' if  x % 2==0 else 'odd'
a(7)
```

# HOF#
Takes one or more functions as arguments, or
Returns a function as a result. or a function that receive another function in input

## Example

```python
def square(x):
    return x**2

def transform(f,L):
    output=[]
    for i in L:
        output.append(f(i))
    print(output)
    
L=[1,2,3,4,5]

transform(square,L)
    
```

```python
L=[1,2,3,4,5]

transform(lambda x: x**2 ,L)
    
```

## Map

The map() function in Python applies a given function to each item of an iterable (such as a list, tuple, etc.) and returns an iterator that yields the results.

function: A function that defines the operation to perform on each element of the iterable.
iterable: The collection of items (e.g., list, tuple) that the function will be applied to

```python
list(map(lambda x: x**2,[1,2,3,4,5]))
```

```python
# odd/ even labeling of list item
L=[2,3,76,89,34,65,11,17]
even_odd=list(map(lambda x : 'even' if x%2==0 else 'odd', L))
print(even_odd)
```

## key difference between map and filter


Both map() and filter() are higher-order functions in Python used to process iterables, but they serve different purposes.

map() Function
Purpose: Applies a given function to every item of an iterable (e.g., list, tuple) and returns an iterator.
    
filter() Function
Purpose: Filters elements from an iterable based on a condition defined by a function. It returns an iterator containing only
those elements for which the function returns True

Function Purpose:

map(): Transforms each element in the iterable based on a function.
filter(): Filters elements from the iterable based on a function (returns only those that satisfy a condition).
Return Type:

Both map() and filter() return an iterator. To see the results, you need to convert them to a list or another collection type.
Function Argument:

map(): The function should return a value that will be included in the result.
filter(): The function should return True or False, determining whether the element should be included in the result.
